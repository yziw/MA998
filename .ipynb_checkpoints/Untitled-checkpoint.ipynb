```julia
# using DynamicalSystems
# using PyPlot
# using Base64
# using PyCall
# @pyimport matplotlib.animation as anim
using Plots
```

```julia
a = 0.9
b = -0.6013
c = 2
d = 0.5
n = 1000
x = zeros(n)
y = zeros(n)
x[1] = -0.5
y[1] = 0.3

for i in 1:n-1
    x[i+1] = x[i]^2 - y[i]^2 + a*x[i] + b*y[i]
    y[i+1] = 2*x[i]*y[i] + c*x[i] + d*y[i]
end

plot(x, y, ls = "none", ms = 1, marker = "o")
```

```julia
a = 0.3
b = 0.6
c = 2
d = 0.27
n = 500
x = zeros(n)
y = zeros(n)
x[1] = 0.1
y[1] = 0.1

for i in 1:n-1
    x[i+1] = x[i]^2 - y[i]^2 + a*x[i] + b*y[i]
    y[i+1] = 2*x[i]*y[i] + c*x[i] + d*y[i]
end
```
