```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = np.linspace(-100, 100, 20)
h = 1
x = np.linspace(-100, 100)
```

```python
plt.figure(figsize = (20, 8))
for i in range(0, len(r)):
    for j in range(0, len(x)):
        y[i, j] = h + r[i]*x[j] - x[j]**2
    plt.plot(x, y[i, :], label = 'r = %.2f' %r[i])
plt.legend()    
    
```

```python
y[1, 1] = h + r
```

```python
r = np.linspace(-15, 15, 20)
x = np.zeros(len(r))
y = np.zeros(len(r))
```

```python
for i in range(0, len(x)):
    x[i] = (r[i] + np.sqrt(r[i]**2 + 4))/2
    y[i] = (r[i] - np.sqrt(r[i]**2 + 4))/2
plt.figure(figsize = (20, 8))
plt.plot(r, x)
plt.plot(r, y)
plt.grid()
```
