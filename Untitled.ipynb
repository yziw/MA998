```julia
# using DynamicalSystems
# using PyPlot
# using Base64
# using PyCall
# @pyimport matplotlib.animation as anim
using Plots
```

```julia
a = 0.9
b = -0.6013
c = 2
d = 0.5
n = 10000
x = zeros(n)
y = zeros(n)
x[1] = 0.1
y[1] = 0.1

for i in 1:n-1
    x[i+1] = x[i]^2 - y[i]^2 + a*x[i] + b*y[i]
    y[i+1] = 2*x[i]*y[i] + c*x[i] + d*y[i]
end
```

```julia
scatter(x, y, ms = 2, marker = "o", markerstrokewidth = 0, legend = :none,
        title = "a = 0.9, b = -0.06013, c = 2, d = 0.5")
xlabel!("x")
ylabel!("y")
```

```julia
gridsize = 500
xs = collect(range(-2, stop = 1, length = gridsize))
ys = collect(range(-2, stop = 1.2, length = gridsize))
resultx = zeros((gridsize, gridsize))
resulty = zeros((gridsize, gridsize));
```

```julia
for i in 1:length(xs)
    for j in 1:length(ys)
        resultx[i, j], resulty[i, j] = ys[j], xs[i]
        for k in 1:100
            xnew = resultx[i, j]^2 - resulty[i, j]^2 + a*resultx[i,j] + b*resulty[i,j]
            ynew = 2*resultx[i, j]*resulty[i, j] + c*resultx[i, j] + d*resulty[i, j]
            resultx[i, j] = xnew
            resulty[i, j] = ynew
        end
    end
end
```

```julia
# figure(figsez = (20, 8))
heatmap(ys, xs, isnan.(resultx), legend = :none, colour = :gray, size = (1000, 700), xlabel = "x", ylabel = "y";
title = "Attractor and trajectory for a = 0.9, b = -0.06013, c = 2, d = 0.5" )
scatter!(x[100:10000], y[100:10000], ms = 2, marker = "o", markercolour = :blue, markerstrokewidth = 0, legend = :none)
```
