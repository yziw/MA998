```julia
using DynamicalSystems
using PyPlot
#using Plots

#### Define the function

function f(dx, x, p, n)
    ## dx is the update rule, x is the array of values
    ## p is that array of parameters
    # a =p[1]
    # b = p[2]
    dx[1] = exp(-p[1]*x[1]^2) + p[2]
    return dx
end

## first entry is function, second is IC, third input is array of parameters
ds = DiscreteDynamicalSystem(f, [0.0], [7.0, 1.0])

a=1

i = 1
pvalues = -1:0.001:1
# ics = [rand() for m in 1:10]
n = 2000
Ttr = 2000
p_index = 2
output = orbitdiagram(ds, i, p_index, pvalues; n = n, Ttr = Ttr)

L = length(pvalues)
x = Vector{Float64}(undef, n*L)
y = copy(x)
for j in 1:L
    x[(1 + (j-1)*n):j*n] .= pvalues[j]
    y[(1 + (j-1)*n):j*n] .= output[j]
end

figure(figsize = (15,8))
#Plots.plot(x,y)
PyPlot.title("total points: $(L*n)")
PyPlot.plot(x, y, ls = "None", ms = 0.2, color = "black", marker = "o", alpha = 0.05)

xlim(pvalues[1], pvalues[end]);
xlabel("b"); ylabel("\$x\$")
tight_layout()
# savefig("q3.pdf")
savefig("q3.png")




f(x, a, b) = exp(-a*x^2) + b
df(x) = -2*a*x*exp(-a*x^2)

a = 7
N = 100000
t = -1:0.01:1
X = zeros(length(t), N)
X[:, 1] = zeros(length(t)).+sqrt(2)/10
i=1
# for b in t
#     for j in 2:N
#         X[i, j] = f(X[i, j-1], a, b)
#     end
#     x = i + 1
# end

for i in 1:length(t)
    for j in 2:N
        X[i, j] = f(X[i, j-1], a, t[i])
    end
end


L = sum(log.(abs.(df.(X))), dims=2)/N



plot(t, L)
axhline(0, color = "r")
```
